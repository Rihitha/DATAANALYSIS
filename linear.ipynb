```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
m=[[5],[7],[10],[20],[22]]
t=[50,45,56,78,34]
LR.fit(m,t)
```

```python
y_pred=LR.predict([[5.5]])
```

```python
#plot the data
import matplotlib.pyplot as plt
plt.scatter(t,m,color='black')
#plot linear regression lilne
#y_pred=LR.predict(5.5)#t=5,7,12,16,20
print(y_pred)
plt.plot(5.5,y_pred,color='blue',linewidth=3,marker='H')
plt.xlabel('Time')
plt.ylabel('Mass')
plt.title('LinearRegression')
plt.show()
```

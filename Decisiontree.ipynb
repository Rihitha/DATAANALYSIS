```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df=pd.read_csv("/content/drive/MyDrive/dataset/demoDT.csv",sep=",")
print(df)
```

```python
feature_cols=['Literacy',' Cleanliness',' Crime_Rate']
feature=df[feature_cols]#features
Target=df.Grade#target variable
```

```python
#Create Decision Tree classifer object
DT=DecisionTreeClassifier()
#train Decision Classifer
DT=DT.fit(feature,Target)
```

```python
#predict the response for test test dataset
l=int(input("enter Literacy:"))
cl=int(input("Enter cleanliness:"))
cr=int(input("enter crimerate"))
pred=DT.predict([[l,cl,cr]])
if pred==1:
      print("Good state")
else:
      print("Bad state")
```

```python
plot_tree(DT)#plot the tarin tree if gini=0.0 it is leaf node
```
